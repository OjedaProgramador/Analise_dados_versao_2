## Tem como sumir com as linhas que são NaN

### Se eu colocar apenas 'dropna' - irá sumir TODAS AS LINHAS QUE TENHAM NaN
### mas eu posso 'refinar' essa busca com: (how= 'all', subset=[('maen', 'No'), ('mean', 'Yes')])

```python
BASE_NOVO.pivot_table(
    index=['JobRole', 'JobLevel'], 
    columns='Attrition',
    values='MonthlyIncome', 
    observed=False,
    aggfunc=['mean', 'size']
    ).dropna(how='all',subset=[('mean', 'No'),('mean', 'Yes')])
```

```python
pd.crosstab(
    [BASE_NOVO['JobRole'], BASE_NOVO['JobLevel']],
    BASE_NOVO['Attrition'],
    normalize='index',
).style.background_gradient(cmap='RdYlGn', axis='columns', vmin=0,vmax=1)
```

## Fazer boxplot automático

```python
job_roles = BASE_NOVO['JobRole'].unique()
job_roles
```

```python
for job in job_roles:
    print(f'JobRole: {job}')
    BASE_NOVO[BASE_NOVO['JobRole'] == job].boxplot(column='MonthlyIncome', by= ['JobLevel','Gender'], grid=False, showmeans=True)
```

## Outra forma de fazer esse processo

```python
for job in job_roles:
    print(f'JobRole: {job}')
    BASE_NOVO[BASE_NOVO['JobRole'] == job].plot(kind='box',column='MonthlyIncome', by= ['JobLevel','Gender'], grid=False, showmeans=True,title=job)
```
