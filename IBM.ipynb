### Soma com o terceiro quartil

    essas operações ocorrem para saber se existem 'outlaier':
    se aparecer valores abaixo de 10,5 ou se aparecer valores acima de 62,5, serão considerados          
        'outlaier'

```python
# esses dados foram retirados da coluna: age (idade)

quartil_1 = 30
quartil_3 = 43
intervalo = 1.5 * (quartil_3 - quartil_1)
print(f'O intervalo interquartil é de {intervalo}')

res = 30 - intervalo
print(res)

res_1 = 43 + intervalo
print(res_1)

```

```python
BASE_NOVO = BASE.drop(['EmployeeCount', 'Over18', 'StandardHours', 'EmployeeNumber'], axis=1)
```

## Maiores valores, usa-se 'BASE_NOVO.nlargest(10, 'Age')'

## Menores valores, usa-se 'BASE_NOVO.nsmallest(10, 'Age')'

```python
BASE_NOVO.nlargest(10, 'Age')
```

```python
BASE_NOVO.nsmallest(10, 'YearsSinceLastPromotion')[['Age','JobRole','YearsSinceLastPromotion', 'JobLevel']]
```

```python
BASE_NOVO.columns
```

```python
colunas_categoricas_nao_ordenada = [
    'BusinessTravel',
    'Department',
    'EducationField',
    'Gender',
    'JobRole',
    'MaritalStatus',
    'OverTime',

]

colunas_categorias_ordenada = [
    'Education',
    'EnvironmentSatisfaction',
    'JobSatisfaction',
    'JobInvolvement',
    'JobLevel',
    'PerformanceRating',
    'RelationshipSatisfaction',
    'StockOptionLevel',
    'WorkLifeBalance',

]

coluna_alvo = ['Attrition']

colunas_numericas = [ coluna for coluna in BASE_NOVO.columns if coluna not in colunas_categoricas_nao_ordenada + colunas_categorias_ordenada + coluna_alvo]
    
```

```python
print(colunas_numericas)
print(colunas_categoricas_nao_ordenada)
```

```python
for coluna in (colunas_categoricas_nao_ordenada + colunas_categorias_ordenada):
    print(f'{coluna}: {BASE_NOVO[coluna].unique()}')
```

```python
for coluna in colunas_categoricas_nao_ordenada:
    BASE_NOVO[coluna] = BASE_NOVO[coluna].astype('category')
BASE_NOVO.info()
```

```python
BASE_NOVO['Attrition'].value_counts().plot(kind='bar', color = ['blue', 'orange']);
```
