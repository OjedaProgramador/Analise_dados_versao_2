```python
import pandas as pd
```

```python
base = 'cereal.csv'

df = pd.read_csv(base)
df
```

```python
df['mfr'].value_counts().plot(
    kind='bar',
    xlabel='Fabricante',
    ylabel= 'Quantidade',
    title='Contagem de cereal por fabricante'
    )
```

```python
df['mfr'].value_counts().plot(
    kind='pie',
    xlabel='Fabricante',
    ylabel= 'Quantidade',
    title='Contagem de cereal por fabricante',
    autopct = '%.1f%%',
    )
```

```python
df['mfr'].value_counts().plot(
    kind='line',
    xlabel='Fabricante',
    ylabel= 'Quantidade',
    title='Contagem de cereal por fabricante'
    )
```
