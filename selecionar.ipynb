### O 'grupby' precisa saber pelo menos duas colunas:
    a principal coluna
    a coluna para ser comparada 
    e precisa colocar uma função para retornar algum valor( soma, média, subtração . . . )

```python
df.groupby('type')['calories'].mean()
```

### Posso colocar mais de uma coluna para agrupar


```python
df.groupby('mfr')[['calories', 'rating']].mean()
```

### Posso reordenar por valores, em ordem crescente ou decrescente !!

```python
df.groupby('mfr')[['calories', 'rating']].mean().sort_values(by='calories', ascending=True) # 'by' traduzido 'por'
```

### O 'describe' pode ser aplicado em uma coluna específica, não somente na tabela inteira

```python
df.groupby('mfr')['calories'].describe()
```

### Agrupar mais de uma coluna principal

#### explicação: As letras (A,G,K,N,P,Q,R) são abreviação dos nomes das empresas. E as letras (H = quente, C = frio)

```python
df.groupby(['mfr', 'type'])['calories'].describe()
```

```python
df['rating'].value_counts(bins=10,normalize=True).sort_index().mul(100).round(2)
```
